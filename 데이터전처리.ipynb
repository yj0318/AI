```python
import os
import shutil
import json
import skimage
import numpy as np
import pandas as pd
from collections import OrderedDict

annoResampleDir = '/content/drive/MyDrive/data_preprocessing/resampling_label/'

annoResampleList = os.listdir(annoResampleDir)
print(len(annoResampleList))
print(annoResampleList[:10])

anno_dict_resample = {"traffic_sign" : 0, "traffic_light" : 0}

for anno in annoResampleList:
    with open(annoResampleDir + anno, encoding="utf-8-sig", errors="ignore") as json_file:
        jsonData = json.load(json_file)
    imgPath = anno.replace(".json", ".jpg")
    df = pd.json_normalize(jsonData)
    filename = df["image.filename"][0]
    #print(df["image.filename"][0])
    for i in range(len(df["annotation"][0])):
        item = df["annotation"][0][i]["class"]
        anno_dict_resample[item] += 1

# 값(value)를 기준으로 정렬한 OrderedDict
ordered_dict = OrderedDict(sorted(anno_dict_resample.items(), key=lambda t:t[1], reverse=True))
print(ordered_dict)
```

```python
# coding:utf-8

import os
import shutil
import json

# 코드를 실행하기 전에 이미지-json 파일이 정확한 쌍을 이루는지 먼저 확인

# annotation 파일들이 있는 경로
annoDir = "/content/drive/MyDrive/data_preprocessing/filter/label_3/"
# image가 있는 경로
imageDir = "/content/drive/MyDrive/data_preprocessing/filter/image_3/"

# anno_dir 내 annotation 파일 이름 리스트업
annoList = os.listdir(annoDir)
print(len(annoList))

# cocoDict에 필요한 images, annotations 정보들의 리스트
imgTmpDict = []
annoTmpDict = []

#class가 traffic_sign 이고, color에 따라 category_id 설정
def set_category_id(annotation):
    color = annotation['color']
    category_id = None

    if color == 'red':
        category_id = 0
    elif color == 'yellow':
        category_id = 1
    elif color == 'blue':
        category_id = 2

    return category_id

idNum = 1
annoIdNum = 60000
for anno in annoList:
    with open(annoDir + anno, encoding="utf-8-sig", errors="ignore") as json_file:
        jsonData = json.load(json_file)

    imgPath = anno.replace(".json", ".jpg")

    imgSize = 1280 * 720

    try :
        images = {
            "license": 1,
            "file_name": jsonData["image"]["filename"],
            "coco_url": "",
            "height": jsonData["image"]["imsize"][1],
            "width": jsonData["image"]["imsize"][0],
            "flickr_url": "",
            "id": idNum
        },
    except :
        images = {
            "license": 1,
            "file_name": jsonData["image"]["filename"],
            "coco_url": "",
            "width": 1280,
            "height": 720,
            "flickr_url": "",
            "id": idNum
        }
    imgTmpDict.append(images)

    for i in range(len(jsonData["annotation"])):
        try: # bbox 정보가 존재할 경우
          bbox = [
            jsonData["annotation"][i]["box"][0],
            jsonData["annotation"][i]["box"][1],
            jsonData["annotation"][i]["box"][2],
            jsonData["annotation"][i]["box"][3]
          ],
          annotation = {
            "segmentation": [
                jsonData["annotation"][i]["box"][0],
                jsonData["annotation"][i]["box"][1]
            ],
            "image_id": idNum,
            "bbox": bbox,
            "class": jsonData["annotation"][i]["class"],
            "category_id": set_category_id(jsonData['annotation'][0]),
            "id": annoIdNum,
          }
        except:
          annotation = {
            "segmentation" : [],
            "image_id": idNum,
            "bbox": [0, 0, 0, 0],
            "class": "traffic_sign",
            "category_id": '',
            "id": annoIdNum,
            }
        annoTmpDict.append(annotation)
    annoIdNum += 1
    idNum += 1


# COCO format annotation이 저장될 dict
cocoDict = {}

cocoDict["info"] = {
  "description": "AI Hub 신호등/도로표지판 인지 영상(수도권)",
  "url": "https://www.aihub.or.kr/aihubdata/data/view.do?currMenu=&topMenu=&aihubDataSe=data&dataSetSn=188",
  "version": "1.3",
  "year": 2021,
  "contributor": "Ha-wook, Jeong",
  "date_created": "2021/11/01",
}

cocoDict["licenses"] = [
  {"url": "https://www.aihub.or.kr/aihubdata/data/view.do?currMenu=&topMenu=&aihubDataSe=data&dataSetSn=188", "id": 1, "name": "라이드플럭스"}
]

cocoDict["images"] = imgTmpDict
cocoDict["annotation"] = annoTmpDict

cocoDict["categories"] = [
  {"supercategory": "warning", "id": 0, "name": "red" },
  {"supercategory": "restriction", "id": 1, "name": "yellow" },
  {"supercategory": "instruction", "id": 2, "name": "blue" },

]

with open("/content/drive/MyDrive/data_preprocessing/filter/3class_coco.json", 'w') as f :
    json.dump(cocoDict, f, ensure_ascii=False)
    print("완료했습니다.")
```

```python
import os
import json
import shutil

# load json and save directory for labels train/val/test
coco_file = '/content/drive/MyDrive/data_preprocessing/filter/3class_coco.json'
save_folder = '/content/drive/MyDrive/data_preprocessing/filter/label_3/'

#source of all the images and destination folder for train/test/val
source_path = "/content/drive/MyDrive/data_preprocessing/filter/image_3/"


# Use os.listdir() to get a list of filenames in the folder
file_names = os.listdir(source_path)

with open(coco_file) as f:
    coco = json.load(f)

images = coco['images']
annotations = coco['annotation']
categories = {cat['id']: cat['name'] for cat in coco['categories']}

os.makedirs(save_folder, exist_ok=True)

for ann in annotations:
    for img in images:
      if(img[0]['id'] == ann['image_id']):
        image = img
    if (image[0]["file_name"] not in file_names):
        continue
    #print(f"image in annotations =   {type(image['id'])}")
    width, height = image[0]['width'], image[0]['height']
    x_min, y_min, x_max, y_max = ann['bbox'][0]
    category_id = ann['category_id']
    filename = image[0]['file_name']
    label_filename = filename.split('.jpg')[0]
    label_path = os.path.join(save_folder, f'{label_filename}.txt')

    with open(label_path, 'a') as f:
          x_center = (x_min + x_max) / (2 * width)
          y_center = (y_min + y_max) / (2 * height)
          bbox_width = (x_max - x_min) / width
          bbox_height = (y_max - y_min) / height

          line = f"{category_id} {x_center} {y_center} {bbox_width} {bbox_height}\n"
          print(line)
          f.write(line)
```
